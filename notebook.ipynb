# Jupyter Notebook

```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
notebook = None
with open('./Demonstration-The-14-3D-Bravais-Lattices-1-0-0-definition.txt', 'r') as file:
    notebook = file.read()

print(notebook)
```

```python
# Extract information from variables (a1cub, a2cub, etc.)
pattern = re.compile(r'(\w+)=\{([^\}]+)\}')
matches = pattern.findall(notebook)

# Visualize variables
for match in matches:
    print(f"{match[0]} = {match[1]}")
```

```python
# Search for the content of the 'coordinates' variable
data = re.findall(r'{([^}]+)}', notebook)

# Function to extract numbers from a string
def extract_numbers(line):
    try:
        return list(map(float, line.split(',')))
    except ValueError:
        return None

# Extract numbers from each line
filtered_data = [extract_numbers(line) for line in data]
filtered_data = [line for line in filtered_data if line is not None]

# Verify that all lines have the same length
valid_data = [line for line in filtered_data if len(line) == 3]

# Create a matrix with the valid data
matrix = np.array(valid_data)

# Create a DataFrame
df = pd.DataFrame(matrix, columns=['x', 'y', 'z'])
print(df)
```

```python

x = df['x']
y = df['y']
z = df['z']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)

ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')

plt.show()
```
